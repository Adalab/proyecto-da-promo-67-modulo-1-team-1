## 🧠 Juego: Trivial de Geografía
### Código del juego
La siguiente celda contiene la lógica del juego. El jugador responde preguntas de geografía escribiendo la respuesta. Gana con 5 aciertos y pierde con 3 fallos.

```python
import random

print('🌍' * 18)
print('🌍¡¡Bienvenido al Trivial de Geografía!!🌍')
print('Responde 5 preguntas bien para ganar')
print('Si fallas 3: pierdes')
print('-' * 40)

# Función para que el jugador pueda introducir palabras en may. o mín., con tildes o sin tildes y para quitar espacios en blanco
def quitar_tildes(texto):
    texto = texto.lower().strip()
    texto = texto.replace('á', 'a')
    texto = texto.replace('é', 'e')
    texto = texto.replace('í', 'i')
    texto = texto.replace('ó', 'o')
    texto = texto.replace('ú', 'u')
    return texto

# Función para ejecutar el juego
def trivial_geografia():

    # Preguntas: decir una respuesta
    concurso = {
        '¿Cuál es el país más grande del mundo?': 'Rusia',
        '¿Cuál es el océano más grande?': 'Océano Pacífico',
        '¿Cuál es la capital de Francia?': 'París',
        '¿Qué río es el más largo del mundo?': 'Amazonas',
        '¿Cuál es el desierto más grande del mundo?': 'Antártico',
        '¿En qué continente está Egipto?': 'África',
        '¿Cuál es la montaña más alta del mundo?': 'Everest',
        '¿Cuál es la capital de Japón?': 'Tokio',
        '¿Qué país tiene más población?': 'China',
        '¿Cuál es el lago más grande del mundo?': 'Mar Caspio',
        '¿En qué continente está Brasil?': 'América del Sur',
        '¿Cuál es la capital de Italia?': 'Roma',
        '¿Qué océano baña las costas de California?': 'Océano Pacífico',
        '¿Cuál es el país más pequeño del mundo?': 'Vaticano',
        '¿Dónde está la Torre Eiffel?': 'París',
        '¿Cuál es el río más largo de Europa?': 'Volga',
        '¿Qué continente es el más grande?': 'Asia',
        '¿Cuál es la capital de Alemania?': 'Berlín',
        '¿Dónde está el Monte Kilimanjaro?': 'Tanzania',
        '¿Qué país tiene forma de bota?': 'Italia',
        '¿Cuál es la capital de España?': 'Madrid',
        '¿Qué mar separa Europa de África?': 'Mar Mediterráneo',
        '¿Cuál es la capital de Canadá?': 'Ottawa',
        '¿Dónde están las pirámides de Giza?': 'Egipto',
        '¿Qué país está al sur de Estados Unidos?': 'México',
        '¿Cuál es la capital de Australia?': 'Canberra',
        '¿Qué continente tiene más países?': 'África',
        '¿Cuál es el río más largo de África?': 'Nilo',
        '¿Dónde está la Gran Muralla?': 'China',
        '¿Cuál es la capital de Argentina?': 'Buenos Aires',
        '¿Qué país es conocido como el país del sol naciente?': 'Japón',
        '¿Cuál es el punto más bajo de la Tierra?': 'Mar Muerto',
        '¿Qué continente está en el hemisferio sur?': 'Antártida',
        '¿Cuál es la capital de Rusia?': 'Moscú',
        '¿Qué isla es la más grande del mundo?': 'Groenlandia',
        '¿Dónde está el desierto del Sahara?': 'África',
        '¿Cuál es la capital de China?': 'Pekín',
        '¿Qué país tiene la mayor cantidad de islas?': 'Suecia',
        '¿Cuál es el volcán más alto del mundo?': 'Ojos del Salado',
        '¿Dónde está el Canal de Panamá?': 'Panamá',
        '¿Cuál es la capital de India?': 'Nueva Delhi',
        '¿Qué cordillera es la más larga del mundo?': 'Los Andes',
        '¿Cuál es la capital de Reino Unido?': 'Londres',
        '¿En que océano está el Polo Norte?': 'Océano Ártico',
        '¿Qué país es famoso por los fiordos?': 'Noruega',
        '¿Cuál es la capital de México?': 'Ciudad de México',
        '¿Qué río pasa por París?': 'Sena',
        '¿Dónde está la selva amazónica?': 'América del Sur',
        '¿Cuál es la capital de Turquía?': 'Ankara',
        '¿Qué país ocupa la mayor parte de la península ibérica?': 'España',
        '¿Cuál es la capital de Grecia?': 'Atenas',
        '¿Qué cordillera separa España de Francia?': 'Los Pirineos',
        '¿Cuál es el país más poblado de Europa?': 'Alemania',
    }

    # Convertimos el diccionario en lista y lo mezclamos; con items iteramos en clave-valor
    lista_concurso = list(concurso.items())
    random.shuffle(lista_concurso)

    # Contadores de aciertos y fallos
    aciertos = 0
    fallos = 0

    # Bucle for para recorrer preguntas y respuestas. Se gana con 5 aciertos y se pierde con 3 fallos
    for pregunta, respuesta in lista_concurso:
        print(pregunta)
        respuesta_jugador = input('Tu respuesta es: ')

        if quitar_tildes(respuesta_jugador) == quitar_tildes(respuesta):
            print('¡Respuesta correcta!')
            aciertos += 1
        else:
            print('¡Incorrecto!')
            print(f'La respuesta correcta es: {respuesta}')
            fallos += 1

        if aciertos == 5:
            print('🏆¡¡¡Enhorabuena: has ganado!!!🏆')
            break
        elif fallos == 3:
            print('😞Has perdido: ¡Inténtalo de nuevo!')
            break
        else:
            print('Continuamos')

    print(f'Juego finalizado. Tienes: Aciertos: {aciertos}, Fallos: {fallos}')

trivial_geografia()
```
