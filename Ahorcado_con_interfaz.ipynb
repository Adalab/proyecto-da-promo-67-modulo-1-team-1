## 🧠 Juego: El Ahorcado
### Código del juego
La siguiente celda contiene la lógica del juego desarrollada por nosotras. El jugador tiene 6 intentos para adivinar una palabra oculta letra a letra. Las palabras pertenecen a 14 categorías temáticas.

```python
import random
import warnings
warnings.filterwarnings("ignore")

def juego_ahorcado():
    palabras = [
    # ROPA
    "camisa","pantalon","falda","vestido","chaqueta","abrigo","zapatos","zapatillas","sandalias","botas",
    "calcetines","gorra","sombrero","bufanda","guantes","cinturon","corbata","sudadera","chaqueton","blusa",

    # TECNOLOGIA
    "ordenador","pantalla","teclado","raton","internet","programacion","python","codigo","software","hardware",
    "archivo","carpeta","servidor","navegador","correo","contraseña","usuario","aplicacion","sistema","datos",

    # ANIMALES
    "perro","gato","caballo","vaca","oveja","cerdo","conejo","tigre","leon","elefante",
    "jirafa","mono","oso","lobo","zorro","aguila","halcon","pajaro","pez","tortuga",

    # COMIDA
    "pizza","hamburguesa","pasta","arroz","pollo","ternera","pescado","ensalada","sopa","pan",
    "queso","huevo","fruta","manzana","platano","naranja","fresa","chocolate","helado","galletas",

    # CIUDADES
    "madrid","barcelona","sevilla","valencia","granada","bilbao","zaragoza","malaga","cordoba","alicante",
    "paris","londres","roma","berlin","lisboa","viena","praga","atenas","amsterdam","bruselas",

    # DEPORTES
    "futbol","baloncesto","tenis","natacion","ciclismo","atletismo","voleibol","rugby","golf","boxeo",
    "esqui","surf","patinaje","karate","judo","yoga","pilates","senderismo","escalada","remo",

    # COLEGIO / UNIVERSIDAD
    "clase","profesor","alumno","examen","apuntes","libro","cuaderno","lapiz","boligrafo","mochila",
    "asignatura","horario","recreo","biblioteca","universidad","instituto","estudio","tarea","trabajo","nota",

    # HOGAR
    "casa","cocina","salon","dormitorio","baño","mesa","silla","sofa","armario","cama",
    "nevera","horno","microondas","lavadora","television","lampara","ventana","puerta","alfombra","espejo",

    # PROFESIONES
    "medico","enfermero","profesor","ingeniero","abogado","arquitecto","programador","diseñador","periodista","fotografo",
    "electricista","fontanero","mecanico","carpintero","panadero","cocinero","camarero","actor","cantante","musico",

    # VIAJES
    "avion","tren","autobus","metro","viaje","hotel","maleta","pasaporte","aeropuerto","estacion",
    "mapa","guia","turista","excursion","playa","montaña","isla","ciudad","pueblo","carretera",

    # NATURALEZA
    "bosque","rio","mar","oceano","lago","montaña","valle","desierto","playa","cascada",
    "nube","lluvia","tormenta","viento","nieve","hielo","arena","roca","arbol","flor",

    # EMOCIONES
    "alegria","tristeza","miedo","enojo","amor","odio","calma","ansiedad","ilusion","esperanza",
    "sorpresa","orgullo","vergüenza","culpa","confianza","pasion","emocion","estres","felicidad","nostalgia",

    # OCIO
    "cine","musica","teatro","libro","serie","pelicula","videojuego","concierto","festival","radio",
    "television","puzzle","ajedrez","cartas","dibujar","pintar","leer","escribir","bailar","cantar",

    # TIEMPO
    "mañana","tarde","noche","ayer","hoy","mañana","semana","mes","año","siglo",
    "hora","minuto","segundo","calendario","reloj","fecha","verano","invierno","primavera","otoño"
]
    palabra_secreta = random.choice(palabras).lower()       
    letras_adivinadas = []
    intentos = 6
    
    # Prepara la palabra a mostrar (guiones)
    palabra_mostrada = ["_"] * len(palabra_secreta)

    print("¡Bienvenido al Juego del Ahorcado!🔥Introduce una letra:")
    print(" ".join(palabra_mostrada))

    while intentos > 0 and "_" in palabra_mostrada:
        letra = input("Ingresa una letra: ").lower()
          
        tildes = {'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u', 'ü': 'u'} # Diccionario para cambiar vocales con tilde a simples
        if letra in tildes:
            letra = tildes[letra]
        if len(letra) != 1 or not letra.isalpha():
            print("Por favor, ingresa una única letra válida.")
            continue

        if len(letra) != 1 or not letra.isalpha():
            print("Por favor, ingresa una única letra válida.")
            continue

        if letra in letras_adivinadas:
            print(f"Ya has intentado la letra '{letra}'.")
        elif letra in palabra_secreta:
            print(f"¡Bien hecho! La letra '{letra}' está en la palabra.")
            # Actualiza la palabra mostrada
            for i, char in enumerate(palabra_secreta):
                if char == letra:
                    palabra_mostrada[i] = letra
        else:
            intentos -= 1 
            print(f"La letra '{letra}' no está en la palabra.")
            if intentos == 5:
                print(f"Vaya comienzo! Como sigas así no sales de esta... \n O")
            if intentos == 4:
                print(f"Has fallado otra vez, estás sin piernas.\n O \n |")
            if intentos == 3:
                print(f"Que haces?! Estás en la cuerda floja! \n  O \n /|")
            if intentos == 2:
                print(f"Cuidado! Vas por mal camino...  \n  O \n /|\ ")
            if intentos == 1:
                print(f"No es por presionarte, pero a la próxima ciao...  \n  O \n /|\ \n / ")
                
        
        letras_adivinadas.append(letra)
        print(" ".join(palabra_mostrada))
        print(f"Intentos restantes: {intentos}")
        print(f"Letras usadas: {', '.join(letras_adivinadas)}")


    if "_" not in palabra_mostrada:
        print(f"\n Enhorabuena! Has adivinado la palabra. \n \U0001F600")

    else:
        print("Te has muerto! Espero que tuvieras tu epitafio preparado...  \n  O \n /|\ \n / \ ")

juego_ahorcado()
```

### Interfaz gráfica
La siguiente celda lanza la interfaz visual del juego. Ejecuta esta celda para jugar con botones en lugar de escribir por consola.

```python
import random
import ipywidgets as w
from IPython.display import display, clear_output

# Estilos inline
ESTILO_TITULO    = 'font-family:Helvetica,Arial,sans-serif; font-size:22px; font-weight:600; color:#1A1A18; margin:0 0 4px;'
ESTILO_SUBTITULO = 'font-family:Helvetica,Arial,sans-serif; font-size:13px; color:#6B6A65; margin:0 0 16px;'
ESTILO_TARJETA   = 'font-family:Helvetica,Arial,sans-serif; background:#fff; border:1px solid #E0DED6; border-radius:10px; padding:16px 20px; margin-bottom:12px; max-width:520px;'
ESTILO_PALABRA   = 'font-family:Courier New,monospace; font-size:28px; font-weight:700; letter-spacing:10px; color:#1A1A18; margin:8px 0;'
ESTILO_INFO      = 'font-family:Helvetica,Arial,sans-serif; font-size:12px; color:#6B6A65; margin:4px 0 10px;'
ESTILO_OK        = 'font-family:Helvetica,Arial,sans-serif; background:#EAF3DE; color:#3B6D11; border-radius:8px; padding:12px 16px; margin-top:8px; max-width:520px;'
ESTILO_ERR       = 'font-family:Helvetica,Arial,sans-serif; background:#FCEBEB; color:#A32D2D; border-radius:8px; padding:12px 16px; margin-top:8px; max-width:520px;'
ESTILO_NEUTRAL   = 'font-family:Helvetica,Arial,sans-serif; background:#F1EFE8; color:#444441; border-radius:8px; padding:12px 16px; margin-top:8px; max-width:520px;'
ESTILO_FIGURA    = 'font-family:Courier New,monospace; font-size:14px; line-height:1.4; background:#F1EFE8; padding:10px 16px; border-radius:8px; max-width:120px; margin-bottom:8px;'

# Fases del ahorcado (0 = ningún fallo, 6 = muerto)
FASES = [
    "",                          # 0 fallos
    "  O",                       # 1 fallo
    "  O\n  |",                  # 2 fallos
    "  O\n /|",                  # 3 fallos
    "  O\n /|\\",                # 4 fallos
    "  O\n /|\\\n /",            # 5 fallos
    "  O\n /|\\\n / \\",         # 6 fallos — muerto
]

TILDES = {'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u', 'ü': 'u'}

PALABRAS = [
    "camisa","pantalon","falda","vestido","chaqueta","abrigo","zapatos","zapatillas","sandalias","botas",
    "calcetines","gorra","sombrero","bufanda","guantes","cinturon","corbata","sudadera","chaqueton","blusa",
    "ordenador","pantalla","teclado","raton","internet","programacion","python","codigo","software","hardware",
    "archivo","carpeta","servidor","navegador","correo","contraseña","usuario","aplicacion","sistema","datos",
    "perro","gato","caballo","vaca","oveja","cerdo","conejo","tigre","leon","elefante",
    "jirafa","mono","oso","lobo","zorro","aguila","halcon","pajaro","pez","tortuga",
    "pizza","hamburguesa","pasta","arroz","pollo","ternera","pescado","ensalada","sopa","pan",
    "queso","huevo","fruta","manzana","platano","naranja","fresa","chocolate","helado","galletas",
    "madrid","barcelona","sevilla","valencia","granada","bilbao","zaragoza","malaga","cordoba","alicante",
    "paris","londres","roma","berlin","lisboa","viena","praga","atenas","amsterdam","bruselas",
    "futbol","baloncesto","tenis","natacion","ciclismo","atletismo","voleibol","rugby","golf","boxeo",
    "esqui","surf","patinaje","karate","judo","yoga","pilates","senderismo","escalada","remo",
    "clase","profesor","alumno","examen","apuntes","libro","cuaderno","lapiz","boligrafo","mochila",
    "asignatura","horario","recreo","biblioteca","universidad","instituto","estudio","tarea","trabajo","nota",
    "casa","cocina","salon","dormitorio","mesa","silla","sofa","armario","cama",
    "nevera","horno","microondas","lavadora","television","lampara","ventana","puerta","alfombra","espejo",
    "medico","enfermero","ingeniero","abogado","arquitecto","programador","periodista","fotografo",
    "electricista","fontanero","mecanico","carpintero","panadero","cocinero","camarero","actor","cantante","musico",
    "avion","tren","autobus","metro","viaje","hotel","maleta","pasaporte","aeropuerto","estacion",
    "mapa","turista","excursion","playa","isla","pueblo","carretera",
    "bosque","rio","mar","oceano","lago","valle","desierto","cascada",
    "nube","lluvia","tormenta","viento","nieve","hielo","arena","roca","arbol","flor",
    "alegria","tristeza","miedo","enojo","amor","calma","ansiedad","ilusion","esperanza",
    "sorpresa","orgullo","culpa","confianza","pasion","emocion","estres","felicidad","nostalgia",
    "cine","musica","teatro","serie","pelicula","videojuego","concierto","festival","radio",
    "puzzle","ajedrez","cartas","dibujar","pintar","leer","escribir","bailar","cantar",
    "tarde","noche","ayer","semana","mes","siglo",
    "hora","minuto","segundo","calendario","reloj","fecha","verano","invierno","primavera","otoño"
]

def lanzar_ahorcado():
    palabra_secreta = random.choice(PALABRAS).lower()
    state = {
        'palabra':    palabra_secreta,
        'mostrada':   ['_'] * len(palabra_secreta),
        'usadas':     [],
        'intentos':   6,
        'bloqueado':  False
    }

    main_out = w.Output()
    display(main_out)

    def refrescar(mensaje_html=''):
        fallos = 6 - state['intentos']
        figura = FASES[fallos].replace('\n', '<br>')
        palabra_html = ' '.join(state['mostrada'])
        usadas_txt = ', '.join(state['usadas']) if state['usadas'] else '—'

        campo  = w.Text(placeholder='Escribe una letra...',
                        layout=w.Layout(width='160px', margin='8px 0'))
        boton  = w.Button(description='Probar letra ✔', button_style='primary',
                          layout=w.Layout(margin='4px 0'))
        result = w.Output()

        if state['bloqueado']:
            campo.disabled = True
            boton.disabled = True

        def intentar(evento):
            if state['bloqueado']:
                return
            letra = campo.value.strip().lower()
            if letra in TILDES:
                letra = TILDES[letra]
            campo.value = ''

            if len(letra) != 1 or not letra.isalpha():
                with result:
                    clear_output(wait=True)
                    display(w.HTML(f'<div style="{ESTILO_NEUTRAL}">⚠️ Introduce una única letra válida.</div>'))
                return

            if letra in state['usadas']:
                with result:
                    clear_output(wait=True)
                    display(w.HTML(f'<div style="{ESTILO_NEUTRAL}">🔁 Ya probaste la letra <strong>{letra}</strong>.</div>'))
                return

            state['usadas'].append(letra)

            if letra in state['palabra']:
                for i, c in enumerate(state['palabra']):
                    if c == letra:
                        state['mostrada'][i] = letra
                msg = f'<div style="{ESTILO_OK}">✅ ¡Bien! La letra <strong>{letra}</strong> está en la palabra.</div>'
            else:
                state['intentos'] -= 1
                msg = f'<div style="{ESTILO_ERR}">❌ La letra <strong>{letra}</strong> no está en la palabra.</div>'

            # ¿Terminó el juego?
            if '_' not in state['mostrada']:
                state['bloqueado'] = True
                campo.disabled = True
                boton.disabled = True
                msg += f'<div style="{ESTILO_OK}"><strong>🎉 ¡Enhorabuena! Has adivinado la palabra. 😀</strong></div>'
                refrescar(msg)
                return
            elif state['intentos'] == 0:
                state['bloqueado'] = True
                campo.disabled = True
                boton.disabled = True
                msg += f'<div style="{ESTILO_ERR}"><strong>💀 ¡Te has muerto! La palabra era: {state["palabra"]}</strong></div>'
                refrescar(msg)
                return

            refrescar(msg)

        boton.on_click(intentar)
        campo.on_submit(intentar)  # también funciona pulsando Enter

        fallos_texto = 6 - state['intentos']
        figura_pre = FASES[fallos_texto]

        with main_out:
            clear_output(wait=True)
            display(w.HTML(f"""
                <p style="{ESTILO_TITULO}">🔥 Juego del Ahorcado</p>
                <p style="{ESTILO_SUBTITULO}">6 intentos para adivinar la palabra oculta</p>
                <div style="{ESTILO_TARJETA}">
                  <p style="{ESTILO_PALABRA}">{' '.join(state['mostrada'])}</p>
                  <pre style="{ESTILO_FIGURA}">{figura_pre}</pre>
                  <p style="{ESTILO_INFO}">
                    💡 Intentos restantes: {state['intentos']}
                    &nbsp;·&nbsp;
                    Letras usadas: {', '.join(state['usadas']) if state['usadas'] else '—'}
                  </p>
                </div>
            """))
            if mensaje_html:
                display(w.HTML(mensaje_html))
            if not state['bloqueado']:
                display(campo, boton)
            else:
                reiniciar = w.Button(description='Jugar de nuevo', button_style='success',
                                     layout=w.Layout(margin='8px 0'))
                reiniciar.on_click(lambda e: lanzar_ahorcado())
                display(reiniciar)
            display(result)

    refrescar()

# Lanzamos el juego
lanzar_ahorcado()
```
