## 🧠 Juego: Piedra, Papel o Tijera
### Código del juego
La siguiente celda contiene la lógica del juego desarrollada por nosotras. El jugador se enfrenta a la máquina en un sistema al mejor de 3: gana quien llegue primero a 2 puntos.

```python
import random

def piedra_papel_tijera():
    juego = ["piedra", "papel", "tijera"]

    puntos_usuario = 0
    puntos_maquina = 0

    while puntos_usuario < 2 and puntos_maquina < 2:
        
        juego_maquina = random.choice(juego)
        juego_usuario = input("piedra papel tijera")
        print(juego_maquina) 

        if juego_usuario == juego_maquina:
            print("empate")
        elif juego_usuario == "piedra" and juego_maquina == "papel":
            print("has perdido, gana papel")
            puntos_maquina = puntos_maquina + 1 
        elif juego_usuario == "tijera" and juego_maquina == "piedra":
            print("has perdido, gana piedra")
            puntos_maquina = puntos_maquina + 1 
        elif juego_usuario == "papel" and juego_maquina == "tijera":
            print("has perdido, gana tijera")
            puntos_maquina = puntos_maquina + 1 
        elif juego_usuario == "piedra" and juego_maquina == "tijera":
            print("has ganado, gana piedra")
            puntos_usuario = puntos_usuario + 1
        elif  juego_usuario == "papel" and juego_maquina == "piedra":
            print("has ganado, gana papel")
            puntos_usuario = puntos_usuario + 1
        elif juego_usuario == "tijera" and juego_maquina == "papel":
            print("has ganado, gana tijera")
            puntos_usuario = puntos_usuario + 1

        if puntos_usuario == 2:
            print("¡¡¡Enhorabuena: GANASTE!!!")          
            break
        elif puntos_maquina == 2:
            print("¡¡¡OHHHH: PERDISTE!!!")
            break
        else:        
            print("Continuamos")   
     
    print(f"Juego finalizado. Puntuacion Consola: {puntos_maquina}, Tu puntuacion: {puntos_usuario}")

piedra_papel_tijera()
```

### Interfaz gráfica
La siguiente celda lanza la interfaz visual del juego. Ejecuta esta celda para jugar con botones en lugar de escribir por consola.

```python
import random
import ipywidgets as w
from IPython.display import display, clear_output

# Estilos inline
ESTILO_TITULO    = 'font-family:Helvetica,Arial,sans-serif; font-size:22px; font-weight:600; color:#1A1A18; margin:0 0 4px;'
ESTILO_SUBTITULO = 'font-family:Helvetica,Arial,sans-serif; font-size:13px; color:#6B6A65; margin:0 0 16px;'
ESTILO_TARJETA   = 'font-family:Helvetica,Arial,sans-serif; background:#fff; border:1px solid #E0DED6; border-radius:10px; padding:16px 20px; margin-bottom:12px; max-width:520px;'
ESTILO_INFO      = 'font-family:Helvetica,Arial,sans-serif; font-size:12px; color:#6B6A65; margin:4px 0 10px;'
ESTILO_OK        = 'font-family:Helvetica,Arial,sans-serif; background:#EAF3DE; color:#3B6D11; border-radius:8px; padding:12px 16px; margin-top:8px; max-width:520px;'
ESTILO_ERR       = 'font-family:Helvetica,Arial,sans-serif; background:#FCEBEB; color:#A32D2D; border-radius:8px; padding:12px 16px; margin-top:8px; max-width:520px;'
ESTILO_NEUTRAL   = 'font-family:Helvetica,Arial,sans-serif; background:#F1EFE8; color:#444441; border-radius:8px; padding:12px 16px; margin-top:8px; max-width:520px;'

EMOJIS = {'piedra': '🪨', 'papel': '📄', 'tijera': '✂️'}
OPCIONES = ['piedra', 'papel', 'tijera']

def lanzar_ppt():
    state = {'puntos_usuario': 0, 'puntos_maquina': 0, 'bloqueado': False}

    main_out = w.Output()
    display(main_out)

    def mostrar_ronda():
        state['bloqueado'] = False

        botones = [
            w.Button(
                description=f"{EMOJIS[op]}  {op.capitalize()}",
                button_style='primary',
                layout=w.Layout(width='130px', margin='4px')
            )
            for op in OPCIONES
        ]
        result = w.Output()

        def jugar(eleccion_usuario):
            if state['bloqueado']:
                return
            state['bloqueado'] = True
            for b in botones:
                b.disabled = True

            eleccion_maquina = random.choice(OPCIONES)

            # Determinar resultado
            if eleccion_usuario == eleccion_maquina:
                resultado = 'empate'
            elif (
                (eleccion_usuario == 'piedra'  and eleccion_maquina == 'tijera') or
                (eleccion_usuario == 'papel'   and eleccion_maquina == 'piedra') or
                (eleccion_usuario == 'tijera'  and eleccion_maquina == 'papel')
            ):
                resultado = 'gana_usuario'
                state['puntos_usuario'] += 1
            else:
                resultado = 'gana_maquina'
                state['puntos_maquina'] += 1

            # Mensajes
            resumen = (
                f"Tú: {EMOJIS[eleccion_usuario]} {eleccion_usuario.capitalize()}  "
                f"· Máquina: {EMOJIS[eleccion_maquina]} {eleccion_maquina.capitalize()}"
            )

            if resultado == 'empate':
                msg_html = f'<div style="{ESTILO_NEUTRAL}">🤝 ¡Empate! — {resumen}</div>'
            elif resultado == 'gana_usuario':
                msg_html = f'<div style="{ESTILO_OK}">✅ ¡Has ganado esta ronda! — {resumen}</div>'
            else:
                msg_html = f'<div style="{ESTILO_ERR}">❌ ¡Ha ganado la máquina! — {resumen}</div>'

            with result:
                clear_output(wait=True)
                display(w.HTML(msg_html))

                if state['puntos_usuario'] >= 2:
                    display(w.HTML(f'<div style="{ESTILO_OK}"><strong>🏆 ¡¡¡Enhorabuena: GANASTE!!!</strong></div>'))
                    mostrar_final()
                elif state['puntos_maquina'] >= 2:
                    display(w.HTML(f'<div style="{ESTILO_ERR}"><strong>😞 ¡¡¡OHHHH: PERDISTE!!!</strong></div>'))
                    mostrar_final()
                else:
                    display(w.HTML(f'<p style="{ESTILO_INFO}">Continuamos... 👇</p>'))
                    sig = w.Button(description='Siguiente ronda →', button_style='primary',
                                   layout=w.Layout(margin='4px 0'))
                    sig.on_click(lambda e: mostrar_ronda())
                    display(sig)

        for b, op in zip(botones, OPCIONES):
            b.on_click(lambda e, o=op: jugar(o))

        with main_out:
            clear_output(wait=True)
            display(w.HTML(f"""
                <p style="{ESTILO_TITULO}">✂️ Piedra, Papel o Tijera</p>
                <p style="{ESTILO_SUBTITULO}">Primero en llegar a 2 puntos gana</p>
                <div style="{ESTILO_TARJETA}">
                  <p style="{ESTILO_INFO}">¿Cuál es tu elección?</p>
                </div>
                <p style="{ESTILO_INFO}">
                  🧑 Tú: {state['puntos_usuario']}
                  &nbsp;·&nbsp;
                  🤖 Máquina: {state['puntos_maquina']}
                </p>
            """))
            display(w.HBox(botones))
            display(result)

    def mostrar_final():
        pu = state['puntos_usuario']
        pm = state['puntos_maquina']
        display(w.HTML(f'<p style="{ESTILO_INFO}">Juego finalizado · Puntuación Consola: {pm} · Tu puntuación: {pu}</p>'))
        reiniciar = w.Button(description='Jugar de nuevo', button_style='success',
                             layout=w.Layout(margin='8px 0'))
        reiniciar.on_click(lambda e: lanzar_ppt())
        display(reiniciar)

    mostrar_ronda()

# Lanzamos el juego
lanzar_ppt()
```
