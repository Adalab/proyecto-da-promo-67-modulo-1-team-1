```python
def piedra_papel_tijera():
    juego = ["piedra", "papel", "tijera"]
    opcion = random.choice(juego)
    juegousuario = input("piedra papel tijera")
    if juegousuario == opcion:
        print("empate")
    elif juegousuario == "piedra" and opcion == "papel":
        print("has perdido, gana piedra")
    elif juegousuario == "piedra" and opcion == "tijera":
        print("has ganado, gana piedra")
    elif  juegousuario == "papel" and opcion == "piedra":
        print("has ganado, gana papel")
    elif juegousuario == "papel" and opcion == "tijera":
        print("has perdido, gana tijera")
    elif juegousuario == "tijera" and opcion == "piedra":
        print("has perdido, gana piedra")
    elif juegousuario == "tijera" and opcion == "papel":
        print("has ganado, gana tijera")

    print(opcion)
```
