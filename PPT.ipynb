```python
import random
```

```python
def piedra_papel_tijera():
    juego = ["piedra", "papel", "tijera"]

    puntos_usuario = 0
    puntos_maquina = 0

    while puntos_usuario < 2 and puntos_maquina <2:
        
        juego_maquina = random.choice(juego)
        juego_usuario = input("piedra papel tijera")
        print(juego_maquina) 

        if juego_usuario == juego_maquina:
            print("empate")
        elif juego_usuario == "piedra" and juego_maquina == "papel":
            print("has perdido, gana papel")
            puntos_maquina = puntos_maquina + 1 
        elif juego_usuario == "tijera" and juego_maquina == "piedra":
            print("has perdido, gana piedra")
            puntos_maquina = puntos_maquina + 1 
        elif juego_usuario == "papel" and juego_maquina == "tijera":
            print("has perdido, gana tijera")
            puntos_maquina = puntos_maquina + 1 
        elif juego_usuario == "piedra" and juego_maquina == "tijera":
            print("has ganado, gana piedra")
            puntos_usuario = puntos_usuario + 1
        elif  juego_usuario == "papel" and juego_maquina == "piedra":
            print("has ganado, gana papel")
            puntos_usuario = puntos_usuario + 1
        elif juego_usuario == "tijera" and juego_maquina == "papel":
            print("has ganado, gana tijera")
            puntos_usuario = puntos_usuario + 1

        if puntos_usuario == 2:
            print("¡¡¡Enhorabuena: GANASTE!!!")          
            break
        elif puntos_maquina == 2:
            print("¡¡¡OHHHH: PERDISTE!!!")
            break
        else:        
            print("Continuamos")   
     
    print(f"Juego finalizado. Puntuacion Consola: {puntos_maquina}, Tu puntuacion: {puntos_usuario}")
```

```python
piedra_papel_tijera()
```
